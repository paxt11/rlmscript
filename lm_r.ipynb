# Linear Modeling in R

```r
df = read.csv("regrex1.csv")
```

```r
head(df)
```

```r
plot(df$x,df$y)
```

```r
model = lm(y~x, data=df)
```

```r
summary_stats= summary(model)
```

```r
summary_stats
```

```r
summary_stats$r.squared
```

```r
coef(model)
```

```r
coef(model)[1]
```

```r
coef(model)[2]
```

```r
y_predict = predict(model, df)
```

```r
plot(df$x,y_predict, xlab = "X axis label", ylab = "Y axis label")
```

```r
plot(df$x,df$y, xlab = "X axis label", ylab = "Y axis label")
lines(df$x, y_predict)
```
